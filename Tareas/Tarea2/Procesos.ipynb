1. **Realiza el Programa 1 que instancie, como clase o con el método Process, 10 procesos. Cada uno de los procesos hijos recibirán un valor entero y un caracter(i, c) enviados por el proceso padre, los procesos escribirán en la salida estándar i veces el caracter c.**

```
import multiprocessing as mp
import random 
import os
import time
```

```
def padre(q,times =10):
  print('Proceso padre:', os.getppid())
  for j in range(times):
      i = random.randint(0,5)
      c = chr(random.randint(97,122))
      q.put([i,c])

def imprime(q):
  print('Proceso hijo:', os.getpid())
  lista = q.get()
  i = lista[0]
  c = lista[1]

  print('La letra',c,'se imprimira',i,'veces:')
  for j in range(i):
    print(c)
```

```
my_queue = mp.Queue()
padre(my_queue)

p1 = mp.Process(target=imprime, args=(my_queue,))
p2 = mp.Process(target=imprime, args=(my_queue,))
p3= mp.Process(target=imprime, args=(my_queue,))
p4= mp.Process(target=imprime, args=(my_queue,))
p5= mp.Process(target=imprime, args=(my_queue,))
p6= mp.Process(target=imprime, args=(my_queue,))
p7= mp.Process(target=imprime, args=(my_queue,))
p8= mp.Process(target=imprime, args=(my_queue,))
p9= mp.Process(target=imprime, args=(my_queue,))
p10= mp.Process(target=imprime,args=(my_queue,))

p1.start()
p2.start()
p3.start()
p4.start()
p5.start()
p6.start()
p7.start()
p8.start()
p9.start()
p10.start()

p1.join()
p2.join()
p3.join()
p4.join()
p5.join()
p6.join()
p7.join()
p8.join()
p9.join()
p10.join()
```
