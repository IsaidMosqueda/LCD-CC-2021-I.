## Implementaci√≥n mejorada

````
import time
import multiprocessing as mp
````

````
def come(tenedor,k):
  tenedor[k]= -1
  tenedor[k+1]= -1

  #Tiempo en comer
  time.sleep(0.5)

  #Terminan de comer despues de un tiempo
  tenedor[k]= 0
  tenedor[k+1]= 0
````

````
def comer(n, filosofos,tenedor):
  for j in range(n):
    for i in range (3):                  
      if (i==2):
        tenedor.append(tenedor[0])

        print("El filosofo", filosofos[i+2],"comera")
        p1 = mp.Process(target=come, args=(tenedor, i+2))
        p1.start()
        p1.join()
        #come(tenedor, i+2)
        #termina_comer(tenedor, i+2)
        print("El fiolosofo", filosofos[i+2], "Ya comio")
        tenedor.pop()
        
      else:
        print("Los filosofos", filosofos[i],' y ',filosofos[i+2],"comeran")
        p1 = mp.Process(target=come, args=(tenedor,i))
        p2 = mp.Process(target=come, args=(tenedor, i+2))
        p1.start()
        p2.start() 
        p1.join()
        p2.join() 
        print("Los filosofos",filosofos[i], "y", filosofos[i+2],  "ya comieron")  
    
    print('')
    print('Todos los filosofos han comido.') 
    print('-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-')                
````

````
import multiprocessing as mp

Filosofos = ['S√≥crates','Platon','Leucipo','Democrito','Arist√≥teles']
tenedor = [0]*len(Filosofos)
n = 5

comer(n,Filosofos,tenedor)
````
